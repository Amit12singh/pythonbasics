```python
import pandas as pd
import numpy as np
```

```python
app=pd.read_excel("loan_sample.xlsx")
app.shape
```

```python
app.info()
```

```python
app.head()
```

```python
app.columns
```

```python
app.pivot_table(index='Profession', aggfunc='count')
```

```python
app.pivot_table(index='Profession', columns='Gender', values='Income', aggfunc='count')
```

```python
app.pivot_table(index='Profession', columns='Gender', values='Income', aggfunc='count',fill_value=0)
```

```python
app.pivot_table(index=['Own_House','Profession'], columns='Gender', values='Income', aggfunc='count',fill_value=0)
```

```python
app.pivot_table(index=['Profession'], columns=['Gender','Own_House'], values='Income', aggfunc='count',fill_value=0)
```

```python
app.pivot_table(index=['Profession'], columns=['Gender','Own_House'], values='Income', aggfunc=['count','sum','mean'],fill_value=0)
```

```python
app.pivot_table(index=['Profession'], columns=['Gender','Own_House'], values='Income', aggfunc=['count'],fill_value=0, margins=True).reset_index()
```

```python
app.pivot_table(index=['Profession'], columns=['Gender','Own_House'], values='Income', aggfunc=['count'],fill_value=0, margins=True,margins_name='Grand_Total').reset_index()
```

```python
app.pivot_table(index=pd.Grouper(key='app_date',freq='Y'),values='Gender',aggfunc='count')
```

```python
app.pivot_table(index=pd.Grouper(key='app_date',freq='M'),values='Gender',aggfunc='count')
```
