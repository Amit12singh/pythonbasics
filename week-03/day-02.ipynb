```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
df=pd.read_csv("bike sharing dataset.csv")
df.head()
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
#display correlation matrix in heatmap
corr = df.corr(numeric_only=True)
plt.figure(figsize=(14,9))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
corr = df.corr(method='kendall', numeric_only=True)
corr
```

```python
corr = df.corr(method='pearson', numeric_only=True)
corr
```

```python
#set up the figure size
plt.figure(figsize=(12,8))
#plot heatmap
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1, vmax=1,
    linewidths=0.5,
    square=True,
    cbar_kws={"shrink":0.8},
)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
features =['temp', 'atemp', 'hum', 'windspeed', 'cnt']
sns.pairplot(df[features ])
plt.suptitle("Pairplot of Bike Sharing Features Colored by Season", y=1.02)
plt.show()
```

```python
sns.scatterplot(data=df, x='temp', y='cnt')
plt.title('Temperature vs count colored')
plt.xlabel('Temperature')
plt.ylabel('Total Count of Rentals')
plt.show()
```

```python
#load tictanic dataset
titanic = sns.load_dataset('titanic')
#Drop missing age values for clarity
titanic = titanic.dropna(subset=['age'])
#Boxplot age by passenger class
plt.figure(figsize=(8,5))
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Age Distribution Across passenger classes')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# Boxplot Age by Sex
plt.figure(figsize=(8, 5))
sns.boxplot(data=titanic, x='sex', y='age')
plt.title('Age Distribution Across Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```
